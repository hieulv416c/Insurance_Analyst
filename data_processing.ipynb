```python
# Danh sách các cột thời gian
date_cols = [
    'TXN_DATE_TIME',
    'POLICY_EFF_DT',
    'LOSS_DT',
    'REPORT_DT',
    'DATE_OF_JOINING'
]

# Convert sang datetime
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# Xem min max
for col in date_cols:
    print(f"\n===== {col} =====")
    print("Min:", df[col].min())
    print("Max:", df[col].max())
```
